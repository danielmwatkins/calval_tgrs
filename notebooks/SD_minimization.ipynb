# Simple rotation algorithm
Let $R_\theta \circ$ be the rotation operator and designate the symmetric difference between shapes $f$ and $g$ as $f \Delta g$. Use $|A|$ to indicate the area of object $A$. The optimum rotation is then $\theta$ such that $|f \Delta (R_\theta \circ g)|$ is minimized:
$$\theta_{opt} = \theta : |f \Delta (R_{\theta} \circ g )| < |f \Delta (R_{\theta_i} \circ g)| \; \forall \; \theta_i \in [\theta_0, \theta_1, \dots, \theta_n]$$

The Julia code below implements a rudimentary version of this search based on rigid rotation with nearest neighbor interpolation over a fixed list of angles. An example of a two step refinement is provided. 

```julia
using Pkg;
Pkg.activate("../scripts/calval") # environment where the needed packages are already installed
using IceFloeTracker
using Plots
using Images
using Interpolations

greaterthan05(x) = x .> 0.5 # used for the image resize step and for binarizing images
imrotate_bin(x, r) = greaterthan05(collect(imrotate(x, deg2rad(r), axes(x), method=BSpline(Constant()))))
imrotate_bin_nocrop(x, r) = greaterthan05(collect(imrotate(x, deg2rad(r), method=BSpline(Constant()))))

# Load the images from file
test_images_loc = "../data/test_images_tracker_setup/"
       
# Add labels and get region properties
# labeled_image = label_components(image);
# props = regionprops_table(labeled_image);
props = []
labeled_images = []
for fname in ["001-baffin_bay-20220911-aqua-labeled_floes.png",
              "022-barents_kara_seas-20060909-aqua-labeled_floes.png"]
    image = load(joinpath(test_images_loc, fname))
        
    # Add labels and get region properties
    labeled_image = label_components(image)
    push!(labeled_images, labeled_image)
    props_table = regionprops_table(labeled_image)
    IceFloeTracker.addfloemasks!(props_table, greaterthan05.(labeled_image))
    push!(props, props_table)        
    # append!(props, IceFloeTracker.addfloemasks!(props_table, greaterthan0.(labeled_image)))
end

all_props = reduce(vcat, props);
```

```julia
"""Pad images by zeros based on the size of the larger of the two images."""
function pad_images(im1, im2)
    max1 = maximum(size(im1))
    max2 = maximum(size(im2))
    
    n = Int64(ceil(maximum([max1, max2])))
    im1_padded = collect(padarray(im1, Fill(0, (n, n), (n, n))))
    im2_padded = collect(padarray(im2, Fill(0, (n, n), (n, n))))
    return im1_padded, im2_padded
end

"""
Align images by selecting and cropping so that r1, c1 and r2, c2 are the center.
These values are expected to be the (integer) centroid of the image. These images should
be already padded so that there is no danger of cutting into the floe shape.
"""
function crop_to_shared_centroid(im1, im2, r1, r2, c1, c2)
    # if r1 == r2 && c1 == c2
    #     return im1, im2
    # end
    
    n1, m1 = size(im1)
    n2, m2 = size(im2)
    new_halfn = minimum([minimum([r1, n1-r1]), minimum([r2, n2-r2])])
    new_halfm = minimum([minimum([c1, m1-c1]), minimum([c2, m2-c2])])

    # check notation: how does julia interpret start and end of array index?
    im1_cropped = im1[1+r1-new_halfn:r1 + new_halfn, 1 + c1-new_halfm:c1 + new_halfm]
    im2_cropped = im2[1+r2-new_halfn:r2 + new_halfn, 1 + c2-new_halfm:c2 + new_halfm]

    # check new sizes: what happens if the new halfm is too big?
    
    return im1_cropped, im2_cropped    
end

"""
Computes the shape difference between im_reference and im_target for each angle (degrees) in test_angles.
The reference image is held constant, while the target image is rotated. The test_angles are interpreted
as the angle of rotation from target to reference, so to find the best match, we rotate the reverse
direction. A perfect match at angle A would imply im_target is the same shape as if im_reference was 
rotated by A degrees.
"""
function shape_difference_rotation(im_reference, im_target, test_angles)
    shape_differences = zeros(size(test_angles))
    init_props = regionprops_table(label_components(im_reference))[1,:] # assumption only one object in image!
    idx = 1
    for angle in test_angles
        im_rotated = imrotate_bin(im_target, -angle)
        rotated_props = regionprops_table(label_components(im_rotated))[1,:]
        im1, im2 = crop_to_shared_centroid(im_reference,
                                           im_rotated,
                                           init_props.row_centroid,
                                           rotated_props.row_centroid,
                                           init_props.col_centroid,
                                           rotated_props.col_centroid)
        # Check here that im1 and im2 sizes are the same
        # Could also add check that the images are nonempty
        # These checks could go inside the crop_to_shared_ccentroid function
        if isequal.(prod(size(im1)), prod(size(im2)))
            a_not_b = im1 .> 0 .&& isequal.(im2, 0)
            b_not_a = im2 .> 0 .&& isequal.(im1, 0)
            shape_differences[idx] = sum(a_not_b .|| b_not_a)
        else
            print("Warning: shapes not equal\n")
            print(angle, size(im1), size(im2), "\n")
            # return im1, im2

            shape_differences[idx] = NaN
        end
        idx += 1
    end
    return shape_differences
end
```

# Testing the minimization
We use manually labeled images as a source of floe shapes for testing.
* How precise can the measured rotation be, given the floe size?
* Can the best fit rotation be found if the floes are very different sizes?

## Example 1: Medium-size floe, manual rotation
We first look at a floe with size 538, rotated by 10 degrees.

```julia
print("Floe size ", all_props[86, :area])
init_floe = all_props[86, :mask]
IceFloeTracker.imshow.(init_floe .* 255)
```

```julia
rotated_floe = imrotate_bin_nocrop(init_floe, 50)
IceFloeTracker.imshow.(rotated_floe .* 255)
```

```julia
init_padded, init_rotated = pad_images(init_floe, rotated_floe)
IceFloeTracker.imshow.([init_padded .* 255, init_rotated .* 255])
```

```julia
# First do a full sweep
test_angles = range(-180, 170, step=10)
sd = shape_difference_rotation(init_padded, init_rotated, test_angles)
first_min = test_angles[argmin(sd)]
print("First guess: ", first_min, "\n")
plot(test_angles, sd, m=:circle)
```

```julia
# Then hone in on minimum
test_angles = range(floor(first_min - 5), ceil(first_min + 5), step = .1)
sd = shape_difference_rotation(init_padded, init_rotated, test_angles)
second_min = test_angles[argmin(sd)]
print("Updated guess: ", second_min)
plot(test_angles, sd, m=:square)
```

The first list of test rotations actually had the true rotation in the list. Attempting refinement with 1/10 degree steps showed that the estimate is noisy. It appears that a reasonably good match can be had from 9.8 to 10.7. Next we look at a larger floe.

## Example 2: Large floe
We next look at floe with area over 1000 pixels.

```julia
print("Floe size ", all_props[12, :area])
init_floe = all_props[12, :mask]
rotated_floe = imrotate_bin_nocrop(init_floe, -85)
init_padded, init_rotated = pad_images(init_floe, rotated_floe)
IceFloeTracker.imshow.([init_padded .* 255, init_rotated .* 255])
```

```julia
# First do a full sweep
test_angles = range(-180, 170, step=10)
sd = shape_difference_rotation(init_padded, init_rotated, test_angles)
first_min = test_angles[argmin(sd)]
print("First guess: ", first_min, "\n")
plot(test_angles, sd, m=:circle)
```

Note the presence of a second minimum at 180 rotation from the true angle. This floe is nearly but not quite symmetric.

```julia
test_angles = range(floor(first_min - 5), ceil(first_min + 5), step = .1)
sd = shape_difference_rotation(init_padded, init_rotated, test_angles)
second_min = test_angles[argmin(sd)]
print("Updated guess: ", second_min)
plot(test_angles, sd, m=:square)
```

The first minimum search was over a 10 degree span. This is not actually a great idea, because we're trying to refine an initial search that was 10 degree spacing. Let's expand that, and use a larger step noting the noisiness of the 1/10 degree steps.

```julia
# Then hone in on minimum
test_angles = range(floor(first_min - 10), ceil(first_min + 10), step = .5)
sd = shape_difference_rotation(init_padded, init_rotated, test_angles)
second_min = test_angles[argmin(sd)]
print("Updated guess: ", second_min)
plot(test_angles, sd, m=:square)
```

We find that the refinement does help -- it's closer this time -- but there's still a 1-degree error.

```julia
# First do a full sweep
test_angles = range(-180, 170, step=5)
sd = shape_difference_rotation(floe1_padded, floe2_padded, test_angles)
first_min = test_angles[argmin(sd)]
print("First guess: ", first_min, "\n")
plot(test_angles, sd, m=:circle)
```

We see that there are two clear minima with similar SD. In the end, this may be good enough -- if it's not the right match, then our shape difference score with the estimated rotation and minimum SD will still be some high number.

```julia
# First do a full sweep
test_angles = range(-180, 170, step=5)
sd = shape_difference_rotation(floe1_padded, floe2_padded, test_angles)
first_min = test_angles[argmin(sd)]
print("First guess: ", first_min, "\n")
plot(test_angles, sd, m=:circle)
```

All the SD values are high, and the minima are not at all convincing. However, it's unlikely that such large area differences would ever get sent to the rotation algorithm, since the ratio tests should filter them out.

# Notes for next steps
* The minimization function may need to be robust to multiple minima, especially in the case that an object has approximate rotational symmetry.
* The angular precision may depend on the floe size. How good is good enough? How can we communicate the uncertainty in the angle estimate? Looking at the size of the trough within 1 or 2 pixels of SD may be enough to get a good estimate.
* We can likely get improved precision if we use antialiasing: using order 1 instead of order 0 splines would mean the test rotations can have partial inclusion in a box. A simple adjustment of the SD calculation to use a sum of the array rather than a sum of booleans would likely help.
* The coarse search then refinement method may only need to be done on floes where the minimum from the coarse search is lower than some threshold. In this case, raw SD should not be used, but rather a normalized SD based on the size of the floes in question.
