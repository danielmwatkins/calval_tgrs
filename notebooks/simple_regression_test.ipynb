```python
import scipy.stats as stats
import numpy as np
```

```python
y = np.array([0.8361110194754405, 
              0.7844221381082916,
              0.7392174204646923,
              0.6933683854146746,
              0.6348451590307866])
x = np.arange(0.1, 0.6, 0.1)
```

```python
stats.linregress(x * 255, y)
```

$ y = -0.49 x + 0.88 $ Reflectance units

$ y = -0.0019 x + 0.88 $ UINT8 units
